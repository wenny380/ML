Реализуем функцию градиентного спуска

```python
import numpy as np

def gradient_descent(X, y, learning_rate, num_iterations, lambda_param):
    # Initialize weight vector w with zeros
    w = np.zeros(X.shape[1])
    
    # Perform gradient descent
    for iteration in range(num_iterations):
        # Compute the model's predictions
        y_pred = np.dot(X, w)
        
        # Compute the gradient of MSE cost function + L2 regularizer
        gradient = 2 * np.dot(X.T, (y_pred - y)) / X.shape[0] + 2 * lambda_param * w
        
        # Update the weights
        w -= learning_rate * gradient
        
    return w
```

```python
import numpy as np

def stochastic_gradient_descent(X, y, learning_rate=0.01, num_epochs=1000):
    _X = X.copy()
    _y = y.copy()
    #m, n = _X.shape # m - количество примеров, n - количество признаков
    #w = np.zeros(n)
    num_features = _X.shape[1]
    num_samples = _X.shape[0]
    
    w = np.random.rand(num_features)  # Инициализация вектора весов w случайными значениями
    
    for epoch in range(num_epochs):
        for i in range(num_samples):
            rand_index = np.random.choice(num_samples)  # Выбор случайного образца из данных
            x_i = _X[rand_index]
            y_i = _y[rand_index]
            
            gradient = np.dot(x_i, np.dot(x_i, w) - y_i)  # Вычисление градиента и обновление весов
            w -= learning_rate * gradient
            
    return w

# X_selected1 = np.random.rand(143, 5)  
# y_train1 = np.random.rand(143) 
# print(X_selected1.shape)
# print(y_train1.shape)
# tol1 = 0.0001
# max_iter1 = 1000
# weights = stochastic_gradient_descent(X_selected1, y_train1, tol1, max_iter1)
```

реализовать метод мини-пакетного градиентного спуска (Mini Batch Gradient Descent)

```python
import numpy as np

def mini_batch_gradient_descent(X, y, learning_rate, num_epochs, batch_size):
    num_samples = X.shape[0]
    num_features = X.shape[1]
    
    w = np.random.rand(num_features)  # Инициализация вектора весов w случайными значениями
    
    for epoch in range(num_epochs):
        indices = np.random.permutation(num_samples)
        X_shuffled = X[indices]
        y_shuffled = y[indices]
        
        for i in range(0, num_samples, batch_size):
            X_batch = X_shuffled[i:i+batch_size]
            y_batch = y_shuffled[i:i+batch_size]
            
            gradient = np.dot(X_batch.T, np.dot(X_batch, w) - y_batch)
            w -= learning_rate * gradient
        
    return w
```
